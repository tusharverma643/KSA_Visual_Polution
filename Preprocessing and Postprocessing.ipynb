```python
import torch
import torchvision
from torchvision.io import read_image
from torchvision.utils import draw_bounding_boxes
from PIL import Image
import matplotlib.pyplot as plt
```

```python
base_path = '/kaggle/input/saudi-dataset-visual-pollution/dataset/images/'
blackList = []
for index,row in df.iterrows():
#     if index not in blackList and index>=19948:
    if index % 800 ==0:
        print(index/200)
    if index not in blackList:
        x = row['image_path']
        img = read_image(base_path + row['image_path'])
        bbox = []
        label = []
        while df.iloc[index,1] == x:
            blackList.append(index)
            row = df.iloc[index,:]
#             print(df.iloc[index,:])
            x_diff = (row['xmax'] - row['xmin'])/2
            y_diff = (row['ymax'] - row['ymin'])/2
            bbox.append([max(row['xmin']*2+x_diff,0), max(row['ymin']*2+y_diff,0), min(row['xmax']*2-x_diff,1920), min(row['ymax']*2-y_diff,1080)])
            label.append(row['name'])
            if index>df.shape[0]:
                break
            index+=1
        
#         bbox = [row['xmin'], row['ymin'], row['xmax'], row['ymax']]
        bbox = torch.tensor(bbox, dtype=torch.int)
#         bbox = bbox.unsqueeze(0)
        img=draw_bounding_boxes(img, bbox, label, width=3, font_size=300)
        img = torchvision.transforms.ToPILImage()(img)
        img = img.save("/kaggle/working/images/" + row['image_path'])
#         plt.imshow(img)
#         torchvision.utils.save_image(img, '/kaggle/working/'+str(row['image_path']))
#         torchvision.utils.save_image(img)
```

```python
df = pd.read_csv('/kaggle/input/saudi-dataset-visual-pollution/dataset/train.csv')
df.head()
```

```python
blackList = []
ctr=0
for index,row in df.iterrows():
    if index not in blackList:
        x = row['image_path']
        file = open('/kaggle/working/labels/'+x[:-4]+'.txt', "w")
        while df.iloc[index,1] == x:
            blackList.append(index)
            row = df.iloc[index,:]
            clas = row['class']
            
            x_diff = (row['xmax'] - row['xmin'])/2
            y_diff = (row['ymax'] - row['ymin'])/2
            l=[max(row['xmin']*2+x_diff,0), max(row['ymin']*2+y_diff,0), min(row['xmax']*2-x_diff,1920), min(row['ymax']*2-y_diff,1080)]
            
            xmin=l[0]/2
            ymin=l[1]/2
            xmax=l[2]/2
            ymax=l[3]/2

            width = (xmax-xmin)/960
            height = (ymax-ymin)/540
            centre_x = (xmax + xmin)/1920
            centre_y = (ymax+ymin)/1080
            file.write(str(clas) + ' ' + str(centre_x) + ' ' + str(centre_y) + ' ' + str(width) + ' ' + str(height) + '\n')
            index+=1
            if index>=df.shape[0]:
                break
        ctr+=1
        file.close()
print(ctr)
import shutil
shutil.make_archive('/kaggle/working/labelZIP', 'zip', '/kaggle/working/labels/')
#         print(text)
# f = open("myfile.txt", "w")
# f.writelines(["Hello World ", "You are welcome to Fcc\n"])
```
